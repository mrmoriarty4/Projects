# Cơ bản về lập trình Python

## Matplotlib

* [Matplotlib](https://matplotlib.org/) là một gói Python phục vụ cho việc vẽ 2D


* Mô đun con của, _matplotlib.pyplot_ chứa nhiều hàm cho phép tạo ra các loại đồ thị khác nhau. 


Ta sẽ bắt đầu với _matplotlib.pyplot_.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Vẽ cơ bản

#### Quy trình

Quy trình chung để vẽ một đường 2D là:

1. Tạo một dãy các giá trị $x$

2. Tạo dãy các gí trị $y$ 

3. Sử dụng lệnh _plt.plot(x,y,[fmt],**kwargs)_ trong đó 

    * [fmt] là chuỗi định dạng (tuỳ chọn)
    
    * **kwargs là đối số để xác định tính chất của đồ thị (tuỳ chọn)

4. Sử dụng các hàm của pyplot để thêm các thành phần khác vào ảnh, chẳng hạn như chú thích hay đường kẻ dạng lưới, v.v. 

5. Dùng lệnh plt.show() để hiển thị ảnh.

Ta bắt đầu bằng một ví dụ đơn giản

```python
x = [-5,-2,0,1,3]
y = [2,-1,1,-4,3]
plt.plot(x,y)
plt.show()
```

Có vài điều cần lưu ý là

* Các dãy $x$ và $y$ tạo thành toạ độ các điểm trên đồ thị

* Đồ thị được xây dựng bằng cách nối các điểm trên bằng những đường thẳng

Nhận xét thứ hai chỉ ra rằng nếu ta muốn vẽ các đường cong trơn thì cần rất nhiều điểm. Ngược lại, đồ thị sẽ không trơn. Chẳng hạn, chúng ta có thể vẽ parabol $y = x^2$ trên đoạn $[-2,2]$ bằng cách chỉ sử dụng 5 điểm

```python
x = [-2,-1,0,1,2]
y = [4,1,0,1,4]
plt.plot(x,y)
plt.show()
```

Ta thấy đường cong không trơn do có quá ít điểm. Để khắc phục ta sẽ tăng số điểm bằng cách sử dụng làm np.linspace() để tạo 100 điểm.

```python
x = np.linspace(-2, 2, 100)
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

Rõ ràng đây là một biểu diễn tốt cho parabol $y = x^2$. Lưu ý rằng số điểm (trong trường hợp này là 100) là tuỳ chọn. Tuy nhiên để có được đường cong trơn ta thường chọn số điểm đủ lớn, phụ thuộc vào hàm muốn vẽ. Đương nhiên nếu số điểm càng lớn thì thời gian để máy tính thực hiện sẽ càng lâu.

#### Thuộc tính của của đường

Đường có một số thuộc tính như màu sắc (color), độ trong suốt (transparency), độ rộng (width) và markers. Chúng ta có thể đặt những thuộc tính này khi gọi hàm plt.plot bằng cách sử dụng các đối số sau 

    Property	                Description
    
    alpha	                   transparency (0.0 transparent through 1.0 opaque)
    
    color (or c)	            any matplotlib color
    
    label	                   text appearing in legend
    
    linestyle (or ls)	       solid, dashed, dashdot, dotted
    
    linewidth (or lw)	       set width of the line
    
    marker	                  set marker style
    
    markeredgecolor (or mec)	any matplotlib color
    
    markerfacecolor (or mfc)	any matplotlib color
    
    markersize (or ms)	      size of the marker

Về màu sắc, ta có thể xác định [matplotlib color](https://matplotlib.org/stable/api/colors_api.html) bằng cách sử dụng tên (__red__, __blue__, ...) hoặc một [RGB](https://www.w3schools.com/colors/colors_rgb.asp) tuple như $(1, 0, 1)$ cho màu purple. Dưới đây chúng ta sẽ vẽ đồ thị của hàm 

$$
    y = e^{-x^2}\cos(2\pi x), \quad x \in [-2, 2].
$$

```python
x = np.linspace(-2,2,41)
y = np.exp(-x**2) * np.cos(2*np.pi*x)
plt.plot(x,y,alpha=0.4,label='Decaying Cosine',
         color='red',linestyle='dashed',linewidth=2,
         marker='o',markersize=5,markerfacecolor='blue',
         markeredgecolor='blue')
plt.ylim([-2,2])
plt.legend()
plt.show()
```

Ở trên chúng ta đã sử dụng plt.legend để hiển thị nhãn của đồ thị (__decaying Cosine__) và plt.ylim để đặt phạm vi của trục tung là khoảng $[-2,2]$.

#### Chuỗi định dạng

Một chuỗi định dạng cho ta một cách ngắn gọn để thêm vào đồ thị các thuộc tính như màu, markers, kiểu đường cong, v.v. Chẳng hạn nếu ta muốn vẽ đồ thị của hàm

$$
    y = \frac{1}{1+x^2}, \quad x \in [-5, 5]
$$

bằng đường đứt nét màu đen ta có thể làm như sau

```python
x = np.linspace(-5, 5, 41)
y = 1/(1 + x**2)
plt.plot(x,y,color = 'black', linestyle = 'dashed', marker = 's')
plt.show()
```

Hoặc ta có thể sử dụng chuỗi định dạng 'ks--' trong đó $k$ ký hiệu cho đường màu đen, $s$ là marker hình vuông và $--$ chỉ dạng đường đứt nét.

```python
x = np.linspace(-5,5,41)
y = 1/(1 + x**2)
plt.plot(x,y,'ks--')
plt.show()
```

### Ví dụ

#### Đa thức Taylor

```python
# Vẽ đồ thị hàm y = cos(x) cũng với đa thức Taylor bậc 2 và bậc 4 của nó

x = np.linspace(-6,6,50)

# Plot y = cos(x)
y = np.cos(x)
plt.plot(x,y,'b',label='cos(x)')

# Plot degree 2 Taylor polynomial
y2 = 1 - x**2/2
plt.plot(x,y2,'r-.',label='Degree 2')

# Plot degree 4 Taylor polynomial
y4 = 1 - x**2/2 + x**4/24
plt.plot(x,y4,'g:',label='Degree 4')

# Add features to our figure
plt.legend()
plt.grid(True,linestyle=':')
plt.xlim([-6,6])
plt.ylim([-4,4])
plt.title('Taylor Polynomials of cos(x) at x=0')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

#### Heart Curve

Vẽ đường cong cho bởi phương trình tham số

$$
    \begin{array}{l}
        x(t) = 16\sin^3(t) \\
        y(t) = 13\cos(t) - 5\cos(2t) - 2\cos(3t) - \cos(4t)
    \end{array}
$$

với $t \in [0, 2\pi]$.

```python
t = np.linspace(0, 2*np.pi, 100)
x = 16*(np.sin(t))**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

# Plot line with RGB tuple (red=1, green=0.2, blue=0.5)
# and 20pt line width

plt.plot(x, y, c = (1, 0.2, 0.5), lw = 20)

# Add features to our figure
plt.title('Heart!')
plt.axis('off')

plt.show()
```

#### Subplots

Hàm plt.subplot nhận ít nhất 3 đầu vào $n$, $m$, $i$ và tạo ra một hình ảnh với một lưới $n \times m$ các subplots và khi đó đặt ô thứ $i$ (đếm dọc theo hàng) làm ô hiện tại.

Ví dụ ta xét [sawtooth wave](https://en.wikipedia.org/wiki/Sawtooth_wave) 

$$
    f(t) = \frac{1}{2} - \frac{1}{\pi}\sum_{k=1}^\infty(-1)^k\frac{\sin(2\pi kt)}{k}
$$

và đặt $f_N(t)$ là tổng riêng thứ $N$ 

$$
    f_N(t) = \frac{1}{2} - \frac{1}{\pi}\sum_{k=1}^N(-1)^k\frac{\sin(2\pi kt)}{k}.
$$

Tạo một lưới $2 \times 2$ các subplots để vẽ 4 tổng riêng đầu tiên

$$ 
    \begin{array}{l} \displaystyle
        f_1(t) = \frac{1}{2} + \frac{\sin(2\pi t)}{\pi} \\ \displaystyle
        f_2(t) = \frac{1}{2} + \frac{\sin(2\pi t)}{\pi} - \frac{\sin(4\pi t)}{2\pi} \\ \displaystyle
        f_3(t) = \frac{1}{2} + \frac{\sin(2\pi t)}{\pi} - \frac{\sin(4\pi t)}{2\pi} + \frac{\sin(6\pi t)}{3\pi} \\ \displaystyle
        f_4(t) = \frac{1}{2} + \frac{\sin(2\pi t)}{\pi} - \frac{\sin(4\pi t)}{2\pi} + \frac{\sin(6\pi t)}{3\pi} + \frac{\sin(8\pi t)}{4\pi}
    \end{array}
$$

```python
t = np.linspace(0,4,200)

fN = 1/2
for N in [1,2,3,4]:
    fN = fN - (-1)**N * np.sin(2*N*np.pi*t)/(N*np.pi)
    plt.subplot(2,2,N)
    plt.plot(t,fN)
    plt.title('N = {}'.format(N))

plt.tight_layout()
plt.show()
```

Xem thêm [tài liệu](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html) về subplots.

### Các dạng đồ thị khác

#### Scatter plots

Một scatter plot có 4 chiều: $x$ coordinate, $y$ coordinate, size và color. Dưới đây ta tạo một scatter plot ngẫu nhiên.

```python
# Set the number of dots in the plot
N = 2000

# Create random x and y coordinates sampled uniformly from [0,1]
x = np.random.rand(N)
y = np.random.rand(N)

# Create random array sampled uniformly from [20,120]
# `size` array is used below to set the size of each dot
size = 100*np.random.rand(N) + 20

# Create random 4-tuples sampled uniformly from [0,1]
# `colors` array is used below to set the color
# (red,green,blue,alpha) of each dot
colors = np.random.rand(N,4)

# Create a figure of size 12 by 5 and create scatter plot
plt.figure(figsize=(12,5))
plt.scatter(x,y,c=colors,s=size)
plt.axis('off')
plt.show()
```

#### Histograms 

Sinh một mảng gồm 10000 số ngẫu nhiên được lấy mẫu từ phân phối chuẩn và tạo một histogram. Ta cũng sẽ vẽ đồ thị của hàm mật độ của phân phối chuẩn

$$
    f(x) = \frac{1}{\sqrt{2\pi}}e^{-x^2/2}.
$$

```python
samples = np.random.randn(10000)
plt.hist(samples,bins=20,density=True,alpha=0.5,color=(0.3,0.8,0.1))
plt.title('Random Samples - Normal Distribution')
plt.ylabel('Frequency')

x = np.linspace(-4,4,100)
y = 1/(2*np.pi)**0.5 * np.exp(-x**2/2)
plt.plot(x,y,'b',alpha=0.8)

plt.show()
```

#### Bar plots

Vẽ bar plot cho [tổng lượng mưa ở Vancouver](https://vancouver.weatherstats.ca/charts/precipitation-monthly.html) theo tháng

```python
month = range(1,13)
precipitation = [98.8,128.8,206.0,138.5,102.2,46.4,1.8,5.0,29.4,114.8,197.0,170.6]

plt.bar(month,precipitation)

plt.xticks(month)
plt.yticks(range(0,300,50))
plt.grid(True,alpha=0.5,linestyle='--')
plt.title('Precipitation in Vancouver, 2017')
plt.ylabel('Total Precipitation (mm)')
plt.xlabel('Month')

plt.show()
```
