# Phương pháp số cho phương trình vi phân (tiếp theo)

## Bài toán biên cho phương trình vi phân

* Trong bài toán giá trị đầu (bài toán Cauchy), phương trình vi phân được nghiên cứu kết hợp với điều kiện liên quan đến giá trị của hàm hoặc giá trị của các đạo hàm tại cùng một giá trị của biến độc lập.

* Đối với các bài toán biên, phương trình vi phân được kết hợp với điều kiện tại các giá trị khác nhau của biến độc lập

* Dạng tổng quát của bài toán biên cho phương trình vi phân cấp hai

$$
    y'' = f(x, y(x), y'(x)), \quad a < x < b, 
$$

$$
    \psi_1(y(a), y(b), y'(a), y'(b)) = 0 \quad \text{và} \quad \psi_2(y(a), y(b), y'(a), y'(b)) = 0
$$

* __Ví dụ:__  

$$
    \begin{cases} y'' = xy + (y')^2 + \sin x, \, 0 < x < 1, \\  \\ y(0) = 1, y(1) = 2\end{cases} \quad \text{hay} \quad \begin{cases} y'' = y^2 + x'e^x \, 1 < x < 4, \\  \\ y'(1) = 0, y(4) = 6\end{cases}
$$

### Phương pháp shooting

* Phương pháp shooting là kỹ thuật chuyển bài toán giá trị biên thành một bài toán giá trị đầu tương đương. Kỹ thuật này có tên gọi là shooting do sự tương tự với quy trình bắn vào một vật thể đứng yên.

![](shooting.png)

* Thủ tục của phương pháp shooting: Để đơn giản, ta xét bài toán biên sau đây

$$
    \begin{cases} y'' = f(x, y, y'), \, a < x < b \\ \\ y(a) = \alpha, \, y(b) = \beta & \end{cases}
$$

_Bước 1:_ Chuyển phương trình bậc hai thành hệ phương trình bậc nhất bằng phép đổi biến $z = y'$
  
$$
    (P_1) \qquad \begin{cases} y' = z & a < x < b \\ z' = f(x, y, z) & a < x < b \\ y(a) = \alpha & \\ z(a) = \xi_0 & \end{cases}
$$

trong đó $\xi_0$ là giá trị đầu của $z$ được chọn dựa vào phán đoán ban đầu.

_Bước 2:_ Sử dụng phương pháp Euler hoặc Runge-Kutta để giải bài toán giá trị đầu cho hệ phương trình bậc nhất ta thu được các giá trị

$$
    y(a)\equiv y(t_0), y(t_1), ...., y(t_N) \equiv y(b),
$$

$$
    z(a)\equiv z(t_0), z(t_1), ...., z(t_N) \equiv z(b).
$$

_Bước 3:_ Tính giá trị $\phi(\xi_0) = y(b) - \beta$. Nếu giá trị đầu $\xi_0$ được chọn đúng thì ta có $\phi(\xi_0) = 0$. Tuy nhiên, thường thì phán đoán ban đầu $\xi_0$ là không tốt nên ta có $\phi(\xi_0) \neq 0$. Lúc này ta giải một hệ phương trình phi tuyến (chẳng hạn dùng phương pháp Newton-Raphson) để tìm một slope $\xi_1$ tốt hơn và lặp lại Bước 2. Quá trình này được thực hiện liên tiếp cho đến khi $\phi(\xi)$ đủ gần 0. 

__Nhận xét:__ Vấn đề cốt lõi là lựa chọn $\xi$ để đảm bảo $\phi(\xi) = 0$. Do đó, dãy giá trị đầu $\xi_0, \xi_1, ...$ có thể được xác định bằng công thức lặp Newton-Raphson

$$
    \xi_{k+1} = \xi_k - \frac{\phi(\xi_k)}{\phi'(\xi_k)}, \,\, k = 0, 1, 2, ...
$$

Ở đây ta lưu ý:

$$
    \phi(\xi) = y(b) - \beta \quad \text{và} \quad \phi'(\xi) = v(b)
$$

trong đó $v$ là nghiệm của bài toán giá trị đầu

$$
    v''(x) = \frac{\partial f}{\partial y}(x, y(x), y'(x))v(x) + \frac{\partial f}{\partial y'}(x, y(x), y'(x))v'(x), \,\, a < x < b,
$$

$$
    v(a) = 0, \quad v'(a) = 1.
$$

Hơn nữa bài toán này cũng có thể viết thành bài toán giá trị đầu cho hệ phương trình vi phân cấp một 

$$
    (P_2) \qquad \begin{cases} v' = u & a < x < b \\ u' = \frac{\partial f}{\partial y}(x, y(x), y'(x))v(x) + \frac{\partial f}{\partial y'}(x, y(x), y'(x))u(x) & a < x < b \\ v(a) = 0 & \\ u(a) = 1 & \end{cases}
$$

__Thuật toán:__

1. Cho trước giá trị đầu $\xi_0$ tuỳ ý và sai số $\epsilon$


2. Giải bài toán giá trị đầu $(P_1)$ và bài toán giá trị đầu $(P_2)$ trên lưới

$$
    x_0 = a < x_1 < \cdots < x_N = b, \quad x_{i} - x_{i-1} = h = \frac{b-a}{N}
$$


3. Với $i = 0, 1, 2, ..., N$ thực hiện các phép tính sau

* $\phi(\xi_i) = y(b) - \beta$, $\quad$ $\phi'(\xi_i) = v(b)$


* Nếu $\big|\phi(\xi_i)\big| < \epsilon$ thì dừng thuật toán. Ngược lại thì tính tiếp


* $\xi_{i+1} = \xi_{i+1} - \phi(\xi_{i})/\phi'(\xi_{i})$


* Quay lại Bước 2 với $\xi_i$ được thay bằng $\xi_{i+1}$.

### Phương pháp sai phân hữu hạn

#### Xấp xỉ đạo hàm

Lý thuyết cơ bản về phép tính vi phân cho ta kết quả sau

__Định lý:__ \, Cho $f : (a, b) \longrightarrow \mathbb{R}$ là hàm đủ tốt. Với $x_0 \in (a, b)$ ta có

$$
    f'(x_0) - \frac{f(x_0+h) - f(x_0)}{h} = \mathcal{O}(h)
$$

$$
    f'(x_0) - \frac{f(x_0+h) - f(x_0 -h)}{2h} = \mathcal{O}(h^2)
$$

và 

$$
    f''(x_0) - \frac{f(x_0+h) + 2f(x_0) + f(x_0 -h)}{h^2} = \mathcal{O}(h^2)
$$

khi $h \to 0$. 

_Nhận xét:_ Các đẳng thức trên dẫn đến các công thức xấp xỉ sai phân cho đạo hàm cấp một và cấp hai tương ứng

$$
    f'(x_0) \approx \frac{f(x_0+h) - f(x_0)}{h}, \quad f'(x_0) \approx \frac{f(x_0+h) - f(x_0 -h)}{2h}, \quad f''(x_0) \approx \frac{f(x_0+h) + 2f(x_0) + f(x_0 -h)}{h^2}
$$

#### Phương pháp sai phân hữu hạn cho bài toán giá trị biên

Ta xét bài toán giá trị biên cho phương trình vi phân cấp hai như sau

$$
    \begin{cases} y'' + p(x)y' + q(x)y = r(x), \, a < x < b \\ \\ y(a) = \alpha, \, y(b) = \beta & \end{cases}
$$

* Chia đoạn $[a, b]$ thành $N$ đoạn con với các điểm chia

$$
    x_0 = a < x_1 < \cdots < b = x_N, \quad x_{i+1} - x_i = h = \frac{b-a}{N}
$$

và mong muốn tính được giá trị của nghiệm tại các nút này: $y(x_i)$ với $i \in \{0, 1, ..., N\}$. Lưu ý là $y(x_0) = \alpha$ và $y(x_1) = \beta$ đã biết.

* Sử dụng công thức sai phân trung tâm cho đạo hàm cấp một và cấp hai

$$
    y'(x_i) \approx \frac{y(x_{i+1}) - y(x_{i-1})}{2h}
$$

$$
    y''(x_i) \approx \frac{y(x_{i+1}) -2y(x_i) + y(x_{i-1})}{h^2}
$$

ta suy ra từ phương trình ban đầu

$$
    \frac{y(x_{i+1}) -2y(x_i) + y(x_{i-1})}{h^2} + p(x_i)\frac{y(x_{i+1}) - y(x_{i-1})}{2h} + q(x_i)y(x_i) = r(x_i), \,\, i = 1, 2, ...., N-1.
$$

Hệ phương trình này có thể viết lại như sau

$$
    \left(1 - \frac{h}{2}p(x_i)\right)y(x_{i-1}) + \left(h^2q(x_i) -2\right)y(x_{i}) + \left(1 + \frac{h}{2}p(x_i)\right)y(x_{i+1}) = h^2r(x_i), \,\, i = 1, 2, ...., N-1.
$$

Đặt $y_i = y(x_i)$ và sử dụng các ký hiệu sau: 

$$
    c_i = 1 - \frac{h}{2}p(x_i), \quad d_i = h^2q(x_i) -2, \quad e_i = 1 + \frac{h}{2}p(x_i), \quad b_i = 
    \begin{cases}
        h^2r(x_i) - \alpha c_i, \, i = 1 \\ 
        \\
        h^2r(x_i), \, 2 \leq i \leq N-2, \\
        \\
        h^2r(x_i) - e_i\beta, \, i = N-1.
    \end{cases}
$$

Khi đó ta thu được hệ phương trình sau

$$
    \begin{bmatrix} 
        d_1 & e_1 & 0 & 0 &  \cdots & 0 \\
        c_2 & d_2 & e_2 & 0 & \cdots & 0 \\
        0 & c_3 & d_3 & e_3 &\cdots & 0 \\
        \vdots & \vdots & \vdots & \vdots & \cdots & \vdots \\
        0 & 0 & 0 & 0 & \cdots & d_{N-1}
    \end{bmatrix}
    \begin{bmatrix} 
        y_1 \\ y_2 \\ y_3 \\ \vdots \\ y_{N-1}
    \end{bmatrix} =
    \begin{bmatrix} 
        b_1 \\ b_2 \\ b_3 \\ \vdots \\ b_{N-1}
    \end{bmatrix}
$$

Giải hệ phương trình này ta sẽ tìm được giá trị của nghiệm tại các nút $x_1, x_2, ..., x_{N-1}$.

__Ví dụ:__ \, Sử dụng phương pháp sai phân hữu hạn để giải bài toán giá trị biên sau

$$
    y'' - 2y = 0, \quad y(0) = 1.2 \quad \text{và} \quad y(1) = 0.9.
$$

Trong trường hợp này ta có $p \equiv 0$, $q \equiv -2$, $r \equiv 0$, $a = 0, \alpha = 1.2$ và $b=1, \beta = 0.9$. 

```python
#h = 0.25

import numpy as np
import matplotlib.pyplot as plt

def p(x):
    return 0
def q(x):
    return -2
def r(x):
    return 0

a = 0
b = 1
alpha = 1.2
beta = 0.9

h = 0.01
N = 101

x = np.arange(0, 1+h, h)
y = np.zeros(N) 
y[0] = alpha
y[N-1] = beta 

c = np.zeros(N-2)
d = np.zeros(N-2)
e = np.zeros(N-2)
b = np.zeros(N-2)
for i in range(1,N-1):
    c[i-1] = 1 - (h/2)*p(x[i])
    d[i-1] = (h**2)*q(x[i]) - 2
    e[i-1] = 1 + (h/2)*p(x[i])
    if (i == 1):
        b[i-1] = (h**2)*r(x[i]) - alpha*c[i-1]
    elif (i == N-2):
        b[i-1] = (h**2)*r(x[i]) - beta*e[i-1]
    else:
        b[i-1] = (h**2)*r(x[i])

A = np.zeros((N-2,N-2))
for i in range(0,N-2):
    for j in range(0,N-2):
        if (i==j):
            A[i][j] = d[i]
        elif (i == j+1):
            A[i][j] = c[i]
        elif (i==j-1):
            A[i][j] = e[i]
        else:
            A[i][j] = 0

from scipy import linalg
sol = linalg.solve(A, b)

for i in range(1,N-1):
    y[i] = sol[i-1]
```

```python
plt.plot(x,y)
plt.show()
```
