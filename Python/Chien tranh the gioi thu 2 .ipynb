```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from collections import Counter
```

```python
url = 'https://vi.wikipedia.org/wiki/Chi%E1%BA%BFn_tranh_th%E1%BA%BF_gi%E1%BB%9Bi_th%E1%BB%A9_hai#C%C3%A1c_n%C6%B0%E1%BB%9Bc_tham_chi%E1%BA%BFn_v%C3%A0_h%E1%BA%ADu_qu%E1%BA%A3'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table', class_ = 'wikitable')
```

```python
table = soup.find('table', class_ = 'wikitable')
```

```python
print(table)
```

```python
table.find_all('th')
```

```python
titles = table.find_all('th')
```

```python
titles_adjusted = [title.text.strip() for title in titles]
titles_adjusted
```

```python
table.find_all('tr')
```

```python
rows_data = table.find_all('tr')
```

```python
rows_data[1]
```

```python
df2 =[]
for row in rows_data[1:]:
  
    individual_row_data = row.find_all('td')
  
    individual_row_data_filled = [data.text.strip() for data in individual_row_data]

    if any(individual_row_data_filled):  # Checks if there's any non-blank element
        df2.append(individual_row_data_filled)
df2
```

```python
df = pd.DataFrame(df2)
df.columns = [titles_adjusted]
df
```
